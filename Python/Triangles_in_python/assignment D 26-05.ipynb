```python
n = 5


for i in range(1, n + 1):
    for j in range(1, i + 1):
        print("*", end=" ")
    
    for j in range(1, 2 * (n - i) + 1):
        print(" ", end=" ")
    
    for j in range(1, i + 1):
        print("*", end=" ")
    
    print()


for i in range(n, 0, -1):
    for j in range(1, i + 1):
        print("*", end=" ")
    
    for j in range(1, 2 * (n - i) + 1):
        print(" ", end=" ")
    
    for j in range(1, i + 1):
        print("*", end=" ")
    
    print()
```

```python
n = 5


for i in range(1, n + 1):
    for j in range(i, n):
        print(" ", end=" ")
    for j in range(1, i + 1):
        print("*", end=" ")
    for j in range(1, i):
        print("*", end=" ")
    print()


for i in range(n - 1, 0, -1):
    for j in range(i, n):
        print(" ", end=" ")
    for j in range(1, i + 1):
        print("*", end=" ")
    for j in range(1, i):
        print("*", end=" ")
    print()
```
